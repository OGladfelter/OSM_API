```python
# https://towardsdatascience.com/loading-data-from-openstreetmap-with-python-and-the-overpass-api-513882a27fd0

import requests
import json
import numpy as np
import matplotlib.pyplot as plt
```

## Bike Parking

```python
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = """
[out:json];
area["ISO3166-1"="US"][admin_level=2];
(node["amenity"="bicycle_parking"](area);
 way["amenity"="bicycle_parking"](area);
 rel["amenity"="bicycle_parking"](area);
);
out center;
"""
response = requests.get(overpass_url, 
                        params={'data': overpass_query})
data = response.json()
```

```python
# Collect coords into list
coords = []
for element in data['elements']:
    if element['type'] == 'node':
        lon = element['lon']
        lat = element['lat']
        coords.append((lon, lat))
    elif 'center' in element:
        lon = element['center']['lon']
        lat = element['center']['lat']
        coords.append((lon, lat))
# Convert coordinates into numpy array
X = np.array(coords)

plt.scatter(X[:, 0], X[:, 1], color='red', alpha=0.25, linewidth = 1, edgecolor='black')
plt.title('Bicycle Parking in USA')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.axis('equal')
plt.xlim(-130,-65)
plt.ylim(20, 60)
#plt.show()
```
